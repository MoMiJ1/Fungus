```python
'''指标为0的人是门将'''

colNames = data.columns[range(15,41)].values

fig, ax = plt.subplots(7,4,figsize=(21,22))
r, c = 0, 0
for i in range(26):
    sns.distplot(data[colNames[i]],fit=norm,kde=True,ax=ax[r,c])
    if c == 3:
        r += 1
        c = 0
    else: c += 1

plt.show()
```

```python
colNames = data.columns[range(41,75)].values

fig, ax = plt.subplots(7,5,figsize=(25,22))
r, c = 0, 0
for i in range(34):
    sns.distplot(data[colNames[i]],fit=norm,kde=True,ax=ax[r,c])
    if c == 4:
        r += 1
        c = 0
    else: c += 1

plt.show()
```

```python
plt.figure(figsize=(25,25))
corrMat = data[data.columns[range(95)]].corr()
sns.heatmap(corrMat,square=True,annot=False,cmap='Blues')
```

```python
'''根据热力图可以看出部分特征有很强的相关性'''

corrMat = data[data.columns[range(15,41)]].corr()
plt.figure(figsize=(18,18))
sns.heatmap(corrMat,annot=True,cmap="rainbow",square=True,fmt=".2f")
```

```python
corrMat = data[data.columns[range(41,75)]].corr()
plt.figure(figsize=(20,20))
sns.heatmap(corrMat,annot=True,cmap="rainbow",square=True,fmt=".2f")
```

```python
corrMat = data.corr()
plt.figure(figsize=(25,23))
sns.heatmap(corrMat,annot=False)
```

```python
color = plt.get_cmap('rainbow')
fig = plt.figure(figsize=(12,10))
k = 12
cols = corrMat.nlargest(k, 'Overall')['Overall'].index
sns.heatmap(data[cols].corr(), annot=True, fmt=".2f", cmap=color,
            yticklabels=cols.values, xticklabels=cols.values)
```
